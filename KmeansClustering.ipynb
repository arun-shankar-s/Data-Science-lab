```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
```

```python
x=iris.data[:,:2]
y=iris.target
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Initialize KMeans with specified parameters
km = KMeans(
    n_clusters=3,            # Number of clusters to form
    init='k-means++',        # Method for initialization of centroids (improves convergence)
    n_init=10,               # Number of times the algorithm will run with different centroid seeds
    max_iter=300,            # Maximum number of iterations per run
    random_state=21,         # Ensures reproducibility of results
    copy_x=True,             # Whether or not to copy the input data (default is True)
    algorithm="auto"         # Automatically selects the best algorithm (default behavior)
)

# Fit the KMeans model to the data (x)
km.fit(x)

# Get the coordinates of the cluster centers
center = km.cluster_centers_

# Print the coordinates of the centroids of the 3 clusters
print("Cluster Centers: \n", center)
```

```python
new_labels=km.labels_
print(new_labels)
print(y)
```

```python
import matplotlib.pyplot as plt

# Create a figure with 1 row and 2 columns of subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Scatter plot for the first subplot (Actual data)
axes[0].scatter(x[:, 0], x[:, 1], c=y, cmap='prism')
axes[0].set_xlabel('Sepal length')
axes[0].set_ylabel('Sepal width')
axes[0].set_title('Actual')

# Scatter plot for the second subplot (Predicted data)
axes[1].scatter(x[:, 0], x[:, 1], c=new_labels, cmap='jet')
axes[1].set_xlabel('Sepal length')
axes[1].set_ylabel('Sepal width')
axes[1].set_title('Predicted')

# Display the plots
plt.show()
```
