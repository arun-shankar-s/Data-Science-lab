```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=[10,20,30,40,50]
x2=[50,40,30,20,10]
labels=["A","B","C","D","E"]
x=np.arange(5)
plt.xticks(x,labels)
plt.bar(x-0.2,x1,color="red",width=0.40)
plt.bar(x+0.2,x2,color="green",width=0.40)
plt.legend(["Men","Women"])
plt.xlabel("Persons")
plt.ylabel("Scores")
plt.title("Persons and Scores")
plt.show()
```
