```python
import numpy as np
```

```python
print("Rank 1 array with 3 elements")
a=np.array([1,2,3])
print(a)
a.shape
```

```python
print("Rank 2 array with 2 rows and 3 coloumn")
a=np.array([[1,2,3],[4,5,6]])
print(a)
a.shape
```

```python
print("Rank 3 array with 1 block, 2 rows and 3 coloumn")
a=np.array([[[1,2,3],[4,5,6]]])
print(a)
a.shape
```

```python
print("Rank 3 array with 2 blocks, 2 rows and 3 coloumn")
a=np.array([
    [[1,2,3],[4,5,6]],
    [[1,2,3],[4,5,6]]
])
print(a)
a.shape
```

```python
print("type",type(a))
```

```python
print("Identity Matrix(4*4)")
a=np.eye(4)
print(a)
```

```python
print("All zeroes array with 2 rows and 3 cols")
a=np.ones((2,3))
print(a)
```

```python
print("All zeroes array with 3 rows and 3 cols")
a=np.ones((3,3))
print(a)
```

```python
print("Random Matrix")
a=np.random.random((2,2,))
print(a)
```

```python
a=np.array([[1,2],[3,4]])
print("Total Sum:",np.sum(a))
```

```python
a=np.array([[1,2],[3,4]])
print("Axis wise sum:",np.sum(a,axis=0))
print("Coordinate wise sum:",np.sum(a,axis=1))
```

```python
a=np.arange(1000)
b=np.arange(1000)
print("Dot Product",np.dot(a,b))
```

```python
print("Element wise Product",np.multiply(a,b))
```
